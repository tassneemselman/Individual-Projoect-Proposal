## Analysis 1 : Conducting a Two-Sample Hypothesis Test for the Difference in Means

This would allow me to test the null hypothesis, which in this case, is that there is no difference in the average levels of depression and loneliness between my two groups (married women with kids and married men with kids). 

**How I would conduct this analysis:**

1. Collect data using set of questions identified above for my depression and variable groups for each of my groups respectively. 
    
2. Then, I would calculate the mean of the depression and loneliness scores for each group. 
    
3. Following this, I would perform a hypothesis test to see if the difference in means is statistically significant.
    
There are two ways to approach my null hypothesis: 

1. There is "no effect" on average: The difference between the population averages of married women with kids and married men with kids is 0.
    
2. The chance that a randomly selected married woman with kids will have a higher level of depression and loneliness than a randomly selected married man with kids is 50/50
    
If the p-value is less than my alpha-significance level, I will reject the null hypothesis, and conclude that there is **evidence** that married women with children experience higher levels of depression and loneliness than married men with children. 

If it is greater than alpha, then I would fail to reject the null hypothesis.

**Limitations:** 

1. I am keeping in mind that correlation does not equal causation, and this only provides evidence.

2. Hypothesis testing cannot tell me why there is a difference, or the magnitude of said difference. 

**Assumptions**

1. I will be choosing an alpha-significance level of 0.05

## Analysis 2 : Construct a Confidence Interval for the Mean Levels of Depression and Loneliness for Each Group

Using this analysis, I will have a range of plausible values for the average levels of depression and loneliness for married women with children and married men with children. 

**How I would conduct my analysis:**

 1. I would use the bootstrapping process to create the confidence intervals. 
    
       a. First step is to use sampling with replacement repeatedly and calculating the mean for each sample. 
        
       b. I would keep track of the mean for each sample using a histogram. 
        
 2. The confidence interval is defined as the range of values that contains the true population mean with a 95% confidence level. 
       a. True population mean in this case is the true mean of depression among my group (Married men with kids or married women with kids) 
        
 3. If the confidence intervals for the two groups don't overlap, this would be evidence to suggest that there is a difference in the mean levels of depression and loneliness between the two groups. 
    
**Limitations**

1. Once again, I am keeping in mind that correlation does not equal causation.
